## trade

```python
trade = pd.read_csv("trade_train.csv")
```

- 6월데이터만

```python
trade_6 = trade[trade['기준년월']==202006]
trade_6
```

```python
group_name = trade['그룹번호'].unique()
group_name
```

### 하나의 그룹 시도

```python
group1 = trade_6[trade_6['그룹번호']==group_name[0]]
group1
```

```python
top3 = group1.sort_values(by=['매수고객수'], ascending=False).head(3)     # top3
top3
```

```python
top3 = trade_6.iloc[[0]] #pd.DataFrame(columns=[df_1.columns.tolist()])
#top3 = top3.drop(0,axis=0) 
#top3

for i in group_name : 
    group1 = trade_6[trade_6['그룹번호']==i]
    top = group1.sort_values(by=['매수고객수'], ascending=False).head(3)     # top3
    top3 = pd.concat([top3,top], axis=0,ignore_index=True)
```

```python
top3 = top3.drop(0,axis=0) 
```

```python
top3 = top3[['그룹번호', '종목번호']]
top3['6월top3여부'] = 1
top3
```
